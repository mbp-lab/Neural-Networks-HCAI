# Classroom Example: The T-Shirt Cannon


## 1-A Shirt Cannon Model: Predicting the Flight Distance Based on Air Pressure

A shirt cannon is a pneumatic device used to launch t-shirts into a crowd during events. The primary mechanism involves compressed air propelling the shirt through a barrel. The distance the shirt travels depends on several factors, including air pressure, the design of the barrel, and the shirt's characteristics.

For the model, we aim to describe how far the shirt flies as a function of air pressure using a simplified physical approach.

### Model Components
The model consists of the following components:

- **Force generated by air pressure**:
  The force exerted by the air pressure ($P$) over the cross-sectional area ($A$) of the barrel is:
  $$
  F = P \cdot A
  $$

- **Acceleration of the shirt**:
  Using Newton's second law, the acceleration ($a$) is given by:
  $$
  a = \frac{F}{m} = \frac{P \cdot A}{m}
  $$
  where $m$ is the mass of the shirt.

- **Velocity at the barrel exit**:
  Assuming the shirt starts at rest, the velocity ($v_0$) at the end of the barrel is determined by the work-energy principle:
  $$
  v_0 = \sqrt{2 \cdot a \cdot L} = \sqrt{\frac{2 \cdot P \cdot A \cdot L}{m}}
  $$
  where $L$ is the barrel's length.

- **Flight distance**:
  The flight distance ($d$) of the shirt in ideal conditions (neglecting air resistance) is determined by projectile motion:
  $$
  d = \frac{v_0^2 \cdot \sin(2\theta)}{g}
  $$
  where $\theta$ is the launch angle and $g$ is the acceleration due to gravity.

This model simplifies the dynamics of the shirt cannon by focusing on the relationships between air pressure, barrel characteristics, and the flight distance. Adjustments for air resistance, shirt deformation, and other real-world factors can refine the predictions.


## A) T-Shirt Cannon Model

This model simulates the range of a t-shirt launched from a cannon, considering two scenarios:
1. **With air resistance**: Accounts for drag due to the t-shirt's cylindrical shape and textile properties.
2. **Without air resistance**: Simplifies motion under ideal projectile physics.

### Assumptions and Parameters

#### General Parameters
- Gravitational acceleration: \( g = 9.81 \, \text{m/s}^2 \)
- Launch angle: \( \theta = 45^\circ \)
- Mass of the t-shirt: \( m = 0.2 \, \text{kg} \)
- Length of the cannon barrel: \( L = 0.5 \, \text{m} \)
- Diameter of the barrel: \( d = 0.1 \, \text{m} \)
- Efficiency of energy transfer: \( \eta = 0.5 \) (50% of pressure energy converted to kinetic energy)

#### Air Resistance
- Drag coefficient for a cylindrical textile object: \( C_d = 1.2 \)
- Air density at sea level: \( \rho = 1.225 \, \text{kg/m}^3 \)


### Physics of the Model

#### Barrel Dynamics
1. **Pressure to Force**:
   \[
   F = P \cdot A
   \]
   where:
   - \( F \) is the force exerted by pressure,
   - \( P \) is the pressure in Pascals (\( \text{psi} \times 6894.76 \)),
   - \( A \) is the cross-sectional area of the barrel: 
     \[
     A = \pi \left( \frac{d}{2} \right)^2
     \]

2. **Acceleration in the Barrel**:
   \[
   a = \frac{F}{m} \cdot \eta
   \]

3. **Velocity at Barrel Exit**:
   Using the kinematic equation for constant acceleration:
   \[
   v_0 = \sqrt{2 \cdot a \cdot L}
   \]

---

#### Trajectory Without Air Resistance
For ideal projectile motion:
\[
R = \frac{v_0^2 \cdot \sin(2\theta)}{g}
\]

#### Trajectory With Air Resistance
The trajectory is computed numerically by simulating the motion under the influence of drag:
1. **Drag Force**:
   \[
   F_d = \frac{1}{2} C_d \cdot \rho \cdot A \cdot v^2
   \]
   where:
   - \( v \) is the speed of the t-shirt at a given moment.

2. **Acceleration Components**:
   \[
   a_x = -\frac{F_d \cdot v_x}{v \cdot m}
   \]
   \[
   a_y = -g - \frac{F_d \cdot v_y}{v \cdot m}
   \]

3. **Numerical Integration**:
   - Positions (\( x, y \)) and velocities (\( v_x, v_y \)) are updated iteratively using small time steps (\( \Delta t \)).

---

### Visualization

The simulation compares the two scenarios:
1. **With Air Resistance** (red solid line): The realistic trajectory affected by drag.
2. **Without Air Resistance** (blue dashed line): The idealized trajectory under simplified conditions.

The plot is constrained to the range of the t-shirt with air resistance to better illustrate the difference. This approach provides an accurate and practical model of a t-shirt cannon, reflecting both real-world physics and idealized conditions.


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # Acceleration due to gravity (m/s^2)
optimal_angle_deg = 45  # Optimal angle for maximum range (degrees)
optimal_angle_rad = np.radians(optimal_angle_deg)
mass_shirt = 0.2  # Mass of the t-shirt (kg)
barrel_length = 0.5  # Length of the cannon barrel (meters)
diameter_barrel = 0.1  # Diameter of the barrel (meters)
efficiency = 0.5  # Efficiency factor (50% of pressure energy converted to kinetic energy)
drag_coefficient = 1.2  # Adjusted drag coefficient for a cylindrical textile object
air_density = 1.225  # Air density at sea level (kg/m^3)

# Define the range function for a given pressure in psi, including air resistance
def range_with_air_resistance(pressure_psi):
    # Convert pressure from psi to force
    pressure_pa = pressure_psi * 6894.76  # Convert psi to Pascals
    area_barrel = np.pi * (diameter_barrel / 2)**2  # Cross-sectional area of the barrel
    force = pressure_pa * area_barrel  # Force exerted by pressure

    # Acceleration in the barrel
    acceleration = (force / mass_shirt) * efficiency  # F = ma, scaled by efficiency

    # Velocity at the end of the barrel
    v0 = np.sqrt(2 * acceleration * barrel_length)  # v^2 = u^2 + 2as, initial velocity u = 0

    # Numerical simulation for motion with air resistance
    dt = 0.01  # Time step (seconds)
    x, y = 0, 0  # Initial position
    vx, vy = v0 * np.cos(optimal_angle_rad), v0 * np.sin(optimal_angle_rad)  # Initial velocity components
    range_x = 0

    while y >= 0:
        # Compute the speed and drag force
        speed = np.sqrt(vx**2 + vy**2)
        drag_force = 0.5 * drag_coefficient * air_density * (diameter_barrel / 2)**2 * np.pi * speed**2

        # Accelerations due to drag and gravity
        ax = -drag_force * (vx / speed) / mass_shirt
        ay = -g - (drag_force * (vy / speed) / mass_shirt)

        # Update velocities
        vx += ax * dt
        vy += ay * dt

        # Update positions
        x += vx * dt
        y += vy * dt

        # Update range
        if y >= 0:
            range_x = x

    return range_x

# Define the range function without air resistance
def range_without_air_resistance(pressure_psi):
    # Convert pressure from psi to force
    pressure_pa = pressure_psi * 6894.76  # Convert psi to Pascals
    area_barrel = np.pi * (diameter_barrel / 2)**2  # Cross-sectional area of the barrel
    force = pressure_pa * area_barrel  # Force exerted by pressure

    # Acceleration in the barrel
    acceleration = (force / mass_shirt) * efficiency  # F = ma, scaled by efficiency

    # Velocity at the end of the barrel
    v0 = np.sqrt(2 * acceleration * barrel_length)  # v^2 = u^2 + 2as, initial velocity u = 0

    # Calculate range using projectile motion formula
    return (v0**2 * np.sin(2 * optimal_angle_rad)) / g

# Simulate pressures from 10 to 100 psi
pressures_psi = np.linspace(10, 100, 500)  # Pressure in psi

# Calculate ranges for each pressure
ranges_with_air = [range_with_air_resistance(p) for p in pressures_psi]
ranges_without_air = [range_without_air_resistance(p) for p in pressures_psi]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(pressures_psi, ranges_with_air, label="With Air Resistance", color="red")
plt.plot(pressures_psi, ranges_without_air, label="Without Air Resistance", linestyle="--", color="blue")
plt.title("T-Shirt Cannon: Range vs Pressure")
plt.xlabel("Pressure (psi)")
plt.ylabel("Range (meters)")
plt.legend()
plt.grid(True)

plt.figure(figsize=(10, 6))
plt.plot(pressures_psi, ranges_with_air, label="With Air Resistance", color="red")
plt.title("T-Shirt Cannon: Range vs Pressure")
plt.xlabel("Pressure (psi)")
plt.ylabel("Range (meters)")
plt.legend()
plt.grid(True)
plt.show()
```

## B) Policy as a Probability Distribution

In reinforcement learning, a **policy** represents how an agent selects actions given a state. When the policy is stochastic, it is modeled as a probability distribution over possible actions. A common choice is a **Gaussian policy** for continuous action spaces.

The Gaussian policy is defined as:

$$
P(a | \mu, \sigma) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left(-\frac{(a - \mu)^2}{2\sigma^2}\right)
$$

### Parameters:
- **$\mu$**: The mean of the distribution, representing the most likely action.
- **$\sigma$**: The standard deviation, controlling the spread (exploration).

### Key Concepts:
1. **Exploration vs Exploitation**:
   - A larger $\sigma$ leads to more exploration, allowing the agent to sample from a wider range of actions.
   - A smaller $\sigma$ narrows the distribution, making the policy more deterministic.

2. **Policy Gradient**:
   The gradient of the log-likelihood with respect to the mean ($\mu$) is:
   $$
   \nabla_\mu \log P(a | \mu, \sigma) = \frac{a - \mu}{\sigma^2}
   $$
   This is used during learning to adjust the policy's parameters, increasing the probability of actions that yield higher rewards.

This probabilistic formulation allows the agent to balance **exploration** (trying new actions) and **exploitation** (choosing known good actions) during training.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gaussian distribution
mean_pressure = 40  # Mean pressure in psi
std_dev_pressure = 10  # Standard deviation in psi

# Generate x values for pressure
pressures_psi = np.linspace(0, 100, 500)

# Gaussian probability distribution (policy)
def gaussian_policy(x, mean, std_dev):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# Gradient of the Gaussian with respect to the mean
def gaussian_gradient_mean(x, mean, std_dev):
    factor = -((x - mean) / (std_dev**2))
    return gaussian_policy(x, mean, std_dev) * factor

# Calculate probabilities and gradients for the pressures
probabilities = gaussian_policy(pressures_psi, mean_pressure, std_dev_pressure)
gradients = gaussian_gradient_mean(pressures_psi, mean_pressure, std_dev_pressure)

# Create the subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the Gaussian policy on the left
axes[0].plot(pressures_psi, probabilities, label=f"Policy (mean={mean_pressure} psi, std={std_dev_pressure} psi)", color="green")
axes[0].set_title("Gaussian Policy for T-Shirt Cannon Pressure")
axes[0].set_xlabel("Pressure (psi)")
axes[0].set_ylabel("Probability Density")
axes[0].legend()
axes[0].grid(True)

# Plot the gradient on the right
axes[1].plot(pressures_psi, gradients, label="Gradient w.r.t. Mean", color="green")
axes[1].set_title("Gradient of Gaussian w.r.t. Mean")
axes[1].set_xlabel("Pressure (psi)")
axes[1].set_ylabel("Gradient Value")
axes[1].legend()
axes[1].grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gaussian distribution
mean_pressure = 40  # Mean pressure in psi
std_dev_pressure = 10  # Standard deviation in psi

# Generate x values for pressure
pressures_psi = np.linspace(0, 100, 500)

# Gaussian probability distribution (policy)
def gaussian_policy(x, mean, std_dev):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# Gradient of the Gaussian with respect to the mean
def gaussian_gradient_mean(x, mean, std_dev):
    factor = -((x - mean) / (std_dev**2))
    return gaussian_policy(x, mean, std_dev) * factor

# Gradient of the log probability with respect to the mean
def log_probability_gradient(x, mean, std_dev):
    return (x - mean) / (std_dev**2)

# Calculate probabilities and gradients for the pressures
probabilities = gaussian_policy(pressures_psi, mean_pressure, std_dev_pressure)
gradients_gaussian = gaussian_gradient_mean(pressures_psi, mean_pressure, std_dev_pressure)
log_probabilities = np.log(probabilities)
gradients_log = log_probability_gradient(pressures_psi, mean_pressure, std_dev_pressure)

# Create the subplots for Gaussian policy and gradient
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the Gaussian policy on the left
axes[0].plot(pressures_psi, probabilities, label=f"Policy (mean={mean_pressure} psi, std={std_dev_pressure} psi)", color="green")
axes[0].set_title("Gaussian Policy for T-Shirt Cannon Pressure")
axes[0].set_xlabel("Pressure (psi)")
axes[0].set_ylabel("Probability Density")
axes[0].legend()
axes[0].grid(True)

# Plot the gradient on the right
axes[1].plot(pressures_psi, gradients_gaussian, label="Gradient of Gaussian", color="blue")
axes[1].set_title("Gradient of Gaussian w.r.t. Mean")
axes[1].set_xlabel("Pressure (psi)")
axes[1].set_ylabel("Gradient Value")
axes[1].legend()
axes[1].grid(True)

# Show the plots
plt.tight_layout()
plt.show()

# Additional figure: Log probabilities and gradient
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the log probabilities on the left
axes[0].plot(pressures_psi, log_probabilities, label="Log of Probabilities", color="blue")
axes[0].set_title("Log of Gaussian Probabilities")
axes[0].set_xlabel("Pressure (psi)")
axes[0].set_ylabel("Log Probability")
axes[0].legend()
axes[0].grid(True)

# Plot the gradient of the log probabilities on the right
axes[1].plot(pressures_psi, gradients_log, label="Gradient of Log Probabilities", color="green")
axes[1].set_title("Gradient of Log Probabilities")
axes[1].set_xlabel("Pressure (psi)")
axes[1].set_ylabel("Gradient Value")
axes[1].legend()
axes[1].grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

## C) Policy Gradient Update

The policy gradient method updates the parameters of a stochastic policy to maximize the expected reward. In our example, the policy is represented by a Gaussian distribution:

$$
P(a | \mu, \sigma) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left(-\frac{(a - \mu)^2}{2\sigma^2}\right)
$$

where:
- $a$: the action (e.g., the sampled pressure).
- $\mu$: the mean of the distribution.
- $\sigma$: the standard deviation.

---

#### Gradient of the Log Probability
The gradient of the log probability with respect to $\mu$ is:

$$
\nabla_\mu \log P(a | \mu, \sigma) = \frac{a - \mu}{\sigma^2}
$$

This term adjusts the mean $\mu$ based on how far the sampled action $a$ is from the current mean, weighted by the variance.

---

#### Policy Gradient Update
The update rule for the mean of the Gaussian policy is:

$$
\Delta \mu = \alpha \cdot R(a) \cdot \nabla_\mu \log P(a | \mu, \sigma)
$$

where:
- $\alpha$: the learning rate (controls the step size of the update).
- $R(a)$: the reward associated with the sampled action $a$.

The updated mean is:

$$
\mu_{\text{new}} = \mu + \Delta \mu
$$

---

#### Explanation
1. The term $\nabla_\mu \log P(a | \mu, \sigma)$ aligns the update direction with the gradient of the probability distribution.
2. The reward $R(a)$ scales the gradient to prioritize actions that yield higher rewards.
3. The learning rate $\alpha$ controls the sensitivity of the update, preventing large jumps in the parameter space.

---

#### Application in the Fourth Plot
In the fourth plot:
- The **blue dashed line** represents the original Gaussian policy.
- The **green solid line** represents the updated Gaussian policy after applying the policy gradient update.
- A **blue arrow** indicates the shift of the mean from $\mu$ to $\mu_{\text{new}}$.

This visualizes how the policy gradient method optimizes the probability distribution to favor actions that lead to higher rewards.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gaussian distribution
mean_pressure = 40  # Mean pressure in psi
std_dev_pressure = 10  # Standard deviation in psi

# Generate x values for pressure
pressures_psi = np.linspace(1, 100, 500)

# Calculate probabilities and gradients for the pressures
probabilities = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - mean_pressure) / std_dev_pressure)**2)
log_probabilities = np.log(probabilities)
gradients_log = (pressures_psi - mean_pressure) / (std_dev_pressure**2)

# Sample an action from the Gaussian policy
sampled_action = np.random.normal(mean_pressure, std_dev_pressure)

# Alias to the existing range_with_air_resistance function
reward_function = range_with_air_resistance

# Compute reward for the sampled action
reward_sampled_action = reward_function(sampled_action)

# Compute gradient of the log probability for the sampled action
log_gradient_sampled_action = (sampled_action - mean_pressure) / (std_dev_pressure**2)

# Plot 1: Gaussian Policy with sampled action
plt.figure(figsize=(7, 5))
plt.plot(pressures_psi, probabilities, label="Gaussian Policy", color="blue")
plt.scatter(sampled_action, 0, color="green", label="Sampled Action", zorder=5)
plt.title("Gaussian Policy and Sampled Action")
plt.xlabel("Pressure (psi)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Reward function with sampled action
plt.figure(figsize=(7, 5))
plt.axvline(mean_pressure, color='green', linestyle='dotted', label='Mean Pressure')
reward_values = [reward_function(p) for p in pressures_psi]
plt.plot(pressures_psi, reward_values, label="Reward Function", color="red")
plt.annotate('', xy=(sampled_action, reward_sampled_action), xytext=(sampled_action, 0), arrowprops=dict(facecolor='green', arrowstyle='->'), zorder=5)
plt.title("Reward Function and Sampled Action")
plt.xlabel("Pressure (psi)")
plt.ylabel("Reward")
plt.ylim(0,75)
plt.legend()
plt.grid(True)
plt.show()

# Plot 3: Log gradient curve with sampled action
plt.figure(figsize=(7, 5))
plt.axvline(mean_pressure, color='green', linestyle='dotted', label='Mean Pressure')
plt.plot(pressures_psi, gradients_log, label="Log Gradient Curve", color="green")
plt.annotate('', xy=(sampled_action, log_gradient_sampled_action), xytext=(sampled_action, 0), arrowprops=dict(facecolor='green', arrowstyle='->'), zorder=5)
plt.title("Log Gradient Curve and Sampled Action")
plt.xlabel("Pressure (psi)")
plt.ylabel("Log Gradient Value")
plt.legend()
plt.grid(True)
plt.show()

# Policy Gradient Update
alpha = 0.5  # Learning rate
mean_update = alpha * reward_sampled_action * log_gradient_sampled_action
new_mean_pressure = mean_pressure + mean_update

# Recalculate probabilities for the updated mean
updated_probabilities = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - new_mean_pressure) / std_dev_pressure)**2)

# Plot 4: Dashed Gaussian Policy
plt.figure(figsize=(7, 5))
plt.plot(pressures_psi, probabilities, label="Gaussian Policy (original one)", linestyle="--", color="blue")
plt.plot(pressures_psi, updated_probabilities, label="Updated Gaussian Policy", color="green")
plt.axvline(new_mean_pressure, color='gray', linestyle='dotted', label='Updated Mean Pressure')
plt.annotate('', xy=(new_mean_pressure, 0), xytext=(mean_pressure, 0), arrowprops=dict(facecolor='blue', arrowstyle='->'), zorder=5, label="Mean Update")
plt.title("Original and Updated Gaussian Policies")
plt.xlabel("Pressure (psi)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Button, VBox
import ipywidgets as widgets

# Parameters for the Gaussian distribution
mean_pressure = 40  # Mean pressure in psi
std_dev_pressure = 10  # Standard deviation in psi

# Generate x values for pressure
pressures_psi = np.linspace(1, 100, 500)

# Calculate probabilities and gradients for the pressures
probabilities = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - mean_pressure) / std_dev_pressure)**2)
log_probabilities = np.log(probabilities)
gradients_log = (pressures_psi - mean_pressure) / (std_dev_pressure**2)

# Alias to the existing range_with_air_resistance function
reward_function = range_with_air_resistance

def interactive_policy(sampled_action):
    # Compute reward for the sampled action
    reward_sampled_action = reward_function(sampled_action)

    # Compute gradient of the log probability for the sampled action
    log_gradient_sampled_action = (sampled_action - mean_pressure) / (std_dev_pressure**2)

    # Policy Gradient Update
    alpha = 0.5  # Learning rate
    mean_update = alpha * reward_sampled_action * log_gradient_sampled_action
    new_mean_pressure = mean_pressure + mean_update

    # Print results
    print(f"Sampled Action: {sampled_action:.2f}")
    print(f"Reward: {reward_sampled_action:.2f}")
    print(f"Log Gradient: {log_gradient_sampled_action:.2f}")
    print(f"Updated Mean: {new_mean_pressure:.2f}")

    # Plot 1: Gaussian Policy with sampled action
    plt.figure(figsize=(6, 6))
    plt.plot(pressures_psi, probabilities, label="Gaussian Policy", color="blue")
    plt.scatter(sampled_action, 0, color="green", label="Sampled Action", zorder=5)
    plt.title("Gaussian Policy and Sampled Action")
    plt.xlabel("Pressure (psi)")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot 2: Reward function with sampled action
    plt.figure(figsize=(6, 6))
    plt.axvline(mean_pressure, color='green', linestyle='dotted', label='Mean Pressure')
    reward_values = [reward_function(p) for p in pressures_psi]
    plt.plot(pressures_psi, reward_values, label="Reward Function", color="red")
    plt.annotate('', xy=(sampled_action, reward_sampled_action), xytext=(sampled_action, 0), arrowprops=dict(facecolor='green', arrowstyle='->'), zorder=5)
    plt.title("Reward Function and Sampled Action")
    plt.xlabel("Pressure (psi)")
    plt.ylabel("Reward")
    plt.ylim(0,75)
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot 3: Log gradient curve with sampled action
    plt.figure(figsize=(6, 6))
    plt.axvline(mean_pressure, color='gray', linestyle='dotted', label='Mean Pressure')
    plt.plot(pressures_psi, gradients_log, label="Log Gradient Curve", color="green")
    plt.annotate('', xy=(sampled_action, log_gradient_sampled_action), xytext=(sampled_action, 0), arrowprops=dict(facecolor='green', arrowstyle='->'), zorder=5)
    plt.title("Log Gradient Curve and Sampled Action")
    plt.xlabel("Pressure (psi)")
    plt.ylabel("Log Gradient Value")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Policy Gradient Update
    alpha = 0.5  # Learning rate
    mean_update = alpha * reward_sampled_action * log_gradient_sampled_action
    new_mean_pressure = mean_pressure + mean_update

    # Recalculate probabilities for the updated mean
    updated_probabilities = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - new_mean_pressure) / std_dev_pressure)**2)

    # Plot 4: Dashed Gaussian Policy
    plt.figure(figsize=(6, 6))
    plt.plot(pressures_psi, probabilities, label="Gaussian Policy (original one)", linestyle="--", color="blue")
    plt.plot(pressures_psi, updated_probabilities, label="Updated Gaussian Policy", color="green")
    plt.axvline(new_mean_pressure, color='gray', linestyle='dotted', label='Updated Mean Pressure')
    plt.annotate('', xy=(new_mean_pressure, 0), xytext=(mean_pressure, 0), arrowprops=dict(facecolor='blue', arrowstyle='->'), zorder=5, label="Mean Update")
    plt.title("Original and Updated Gaussian Policies")
    plt.xlabel("Pressure (psi)")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Interactive slider and button
slider = FloatSlider(value=mean_pressure, min=0, max=100, step=1, description="Sample Action")

# Function to draw a sample and update slider
def draw_sample(_):
    slider.value = np.random.normal(mean_pressure, std_dev_pressure)

# Create a button to draw a sample
sample_button = Button(description="Draw Sample")
sample_button.on_click(draw_sample)

# Display the slider and button
ui = VBox([sample_button])
interact(interactive_policy, sampled_action=slider)
ui
```

### Progression of Learning

Shown is in the following plot the subsequent application of policy gradient update steps.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gaussian distribution
mean_pressure = 40  # Initial mean pressure in psi
std_dev_pressure = 10  # Standard deviation in psi

# Generate x values for pressure
pressures_psi = np.linspace(1, 100, 500)

# Calculate initial probabilities for the pressures
probabilities = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - mean_pressure) / std_dev_pressure)**2)

# Alias to the existing range_with_air_resistance function
reward_function = range_with_air_resistance

# Initialize variables
samples = []  # Store all sampled actions
mean_values = [mean_pressure]  # Store the mean values at each step
updated_mean_pressure = mean_pressure  # Start with initial mean

update_steps=100

# Iterate for 20 steps
for _ in range(update_steps):
    # Sample an action from the current Gaussian policy
    sampled_action = np.random.normal(updated_mean_pressure, std_dev_pressure)
    samples.append(sampled_action)

    # Compute reward for the sampled action
    reward_sampled_action = reward_function(sampled_action)

    # Compute gradient of the log probability for the sampled action
    log_gradient_sampled_action = (sampled_action - updated_mean_pressure) / (std_dev_pressure**2)

    # Policy Gradient Update
    alpha = 0.1  # Learning rate
    mean_update = alpha * reward_sampled_action * log_gradient_sampled_action
    updated_mean_pressure += mean_update

    # Store the updated mean
    mean_values.append(updated_mean_pressure)

# Recalculate probabilities for the updated mean
updated_probabilities = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - updated_mean_pressure) / std_dev_pressure)**2)

# Combined Plot 1: Initial Gaussian Policy, all 20 samples, and final updated Gaussian Policy
plt.figure(figsize=(10, 6))
plt.plot(pressures_psi, probabilities, label="Initial Gaussian Policy", linestyle="--", color="blue")
plt.plot(pressures_psi, updated_probabilities, label="Updated Gaussian Policy (20 Steps)", color="green")
plt.scatter(samples, [0] * len(samples), color="red", label="Sampled Actions", zorder=5)
plt.title("Initial and Updated Gaussian Policies with Sampled Actions")
plt.xlabel("Pressure (psi)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

# Combined Plot 2: Mean over iterations and sampled actions
plt.figure(figsize=(10, 6))
plt.plot(range(update_steps+1), mean_values, label="Mean Values (Green Line)", color="green")
plt.scatter(range(1, update_steps+1), samples, color="red", label="Sampled Actions", zorder=5)
plt.title("Mean Updates and Sampled Actions Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Pressure (psi)")
plt.legend()
plt.grid(True)
plt.show()
```

## D) Using Advantage Estimation

Above, probabilities of the policies were always updated towards a sample when this provided any positive reward (and it was scaled simply by the reward). We can observe some oscillations, as four our example also values lower than the current mean still have a high reward and appear attractive. 

Now, over time we realize that these values in our trivial example are not performing that well --- we have seen many rewards and can compare these to each other. We can introduce the same into our policy update by replacing the reward itself with an advantage function: How does the current earned reward compare to the mean reward we have experienced?

In the provided implementation, Policy 2 incorporates elements of the **actor-critic framework** by using advantage-based updates:

#### Key Concepts:
1. **Actor**:
   - The policy is represented as a Gaussian distribution:
     $$
     P(a | \mu, \sigma) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left(-\frac{(a - \mu)^2}{2\sigma^2}\right)
     $$
   - The mean, $\mu$, is updated to maximize the expected reward.

2. **Critic**:
   - The critic estimates the baseline value function using the **mean reward** over time:
     $$
     \text{Mean Reward}_{t+1} = \frac{t \cdot \text{Mean Reward}_t + R(a_t)}{t + 1}
     $$
   - This baseline is used to compute the **advantage**:
     $$
     A(a_t) = R(a_t) - \text{Mean Reward}_{t}
     $$

3. **Policy Update**:
   - The policy gradient step updates $\mu$ using the advantage:
     $$
     \Delta \mu = \alpha \cdot A(a_t) \cdot \nabla_\mu \log P(a_t | \mu, \sigma)
     $$
   - Here, $\nabla_\mu \log P(a_t | \mu, \sigma)$ is:
     $$
     \frac{a_t - \mu}{\sigma^2}
     $$

#### Implementation Details:
- **Policy 1**: Standard policy gradient using raw rewards.
- **Policy 2**: Advantage-based policy gradient normalizing rewards with a mean baseline to reduce variance in updates.

This approach demonstrates how advantage estimation improves stability and learning efficiency in reinforcement learning scenarios.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gaussian distributions
mean_pressure1 = 40  # Initial mean pressure for policy 1 in psi
mean_pressure2 = 40  # Initial mean pressure for policy 2 in psi
std_dev_pressure = 10  # Standard deviation in psi

# Generate x values for pressure
pressures_psi = np.linspace(1, 100, 500)

# Calculate initial probabilities for the pressures
probabilities1 = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - mean_pressure1) / std_dev_pressure)**2)
probabilities2 = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - mean_pressure2) / std_dev_pressure)**2)

# Alias to the existing range_with_air_resistance function
reward_function = range_with_air_resistance

# Initialize variables for both policies
samples1 = []  # Store sampled actions for policy 1
samples2 = []  # Store sampled actions for policy 2
mean_values1 = [mean_pressure1]  # Store mean values for policy 1
mean_values2 = [mean_pressure2]  # Store mean values for policy 2
updated_mean_pressure1 = mean_pressure1  # Start with initial mean for policy 1
updated_mean_pressure2 = mean_pressure2  # Start with initial mean for policy 2
mean_rewards = []  # Store mean rewards for policy 2

# Learning rates
alpha_1 = 0.2
alpha_2 = 1.0
update_steps = 100

# Iterate for 100 steps
for step in range(update_steps):
    # Policy 1: Sample, update, and record
    sampled_action1 = np.random.normal(updated_mean_pressure1, std_dev_pressure)
    samples1.append(sampled_action1)
    reward_sampled_action1 = reward_function(sampled_action1)
    log_gradient_sampled_action1 = (sampled_action1 - updated_mean_pressure1) / (std_dev_pressure**2)
    mean_update1 = alpha_1 * reward_sampled_action1 * log_gradient_sampled_action1  # Learning rate for policy 1
    updated_mean_pressure1 += mean_update1
    mean_values1.append(updated_mean_pressure1)

    # Policy 2: Sample, update, and record
    sampled_action2 = np.random.normal(updated_mean_pressure2, std_dev_pressure)
    samples2.append(sampled_action2)
    reward_sampled_action2 = reward_function(sampled_action2)

    # Track mean reward
    mean_rewards.append(reward_sampled_action2 if step == 0 else (mean_rewards[-1] * step + reward_sampled_action2) / (step + 1))

    # Compute advantage (reward - mean reward)
    advantage = reward_sampled_action2 - mean_rewards[-1]

    log_gradient_sampled_action2 = (sampled_action2 - updated_mean_pressure2) / (std_dev_pressure**2)
    mean_update2 = alpha_2 * advantage * log_gradient_sampled_action2  # Learning rate for policy 2
    updated_mean_pressure2 += mean_update2
    mean_values2.append(updated_mean_pressure2)

# Recalculate probabilities for the updated means
updated_probabilities1 = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - updated_mean_pressure1) / std_dev_pressure)**2)
updated_probabilities2 = (1 / (std_dev_pressure * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((pressures_psi - updated_mean_pressure2) / std_dev_pressure)**2)

# Combined Plot 1: Initial Gaussian Policies, all samples, and final updated Gaussian Policies
plt.figure(figsize=(10, 6))
plt.plot(pressures_psi, probabilities1, label="Initial Gaussian Policy 1", linestyle="--", color="blue")
plt.plot(pressures_psi, updated_probabilities1, label="Updated Gaussian Policy 1", color="green")
plt.scatter(samples1, [0] * len(samples1), color="red", label="Sampled Actions Policy 1", zorder=5)
plt.plot(pressures_psi, probabilities2, label="Initial Gaussian Policy 2", linestyle="--", color="purple")
plt.plot(pressures_psi, updated_probabilities2, label="Updated Gaussian Policy 2", color="blue")
plt.scatter(samples2, [0] * len(samples2), color="orange", label="Sampled Actions Policy 2", zorder=5)
plt.title("Initial and Updated Gaussian Policies with Sampled Actions")
plt.xlabel("Pressure (psi)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

# Combined Plot 2: Mean over iterations and sampled actions for both policies
plt.figure(figsize=(10, 6))
plt.plot(range(update_steps+1), mean_values1, label="Mean Values Policy 1", color="green")
plt.scatter(range(1, update_steps+1), samples1, color="red", label="Sampled Actions Policy 1", zorder=5)
plt.plot(range(update_steps+1), mean_values2, label="Mean Values Policy 2", color="blue")
plt.scatter(range(1, update_steps+1), samples2, color="orange", label="Sampled Actions Policy 2", zorder=5)
plt.title("Mean Updates and Sampled Actions Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Pressure (psi)")
plt.legend()
plt.grid(True)
plt.show()
```
